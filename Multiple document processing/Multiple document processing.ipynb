## Fruits and Veggies

The US Department of Agriculture publishes price estimates for fruits and vegetables [online](https://www.ers.usda.gov/data-products/fruit-and-vegetable-prices/fruit-and-vegetable-prices/). The most recent estimates are based on a 2013 survey of US retail stores.

The estimates are provided as a collection of MS Excel files, with one file per fruit or vegetable. The `assignment3_data.zip` file contains the fruit and vegetable files in the directories `fruit` and `vegetables`, respectively.

```python
import pandas as pd
import numpy as np
import os

#get files names from fruit fold and store the file names 
files = os.listdir('assignment3_data/fruit')

#set the column name for fruit data frame
maincolumns = ['type','food','form','price_per_lb','yield','lb_per_cup','price_per_cup']

#set a blank dataframe for fruit information
fruit = pd.DataFrame(columns=maincolumns)


for f in files:
    #read each file corespondingly
    df = pd.read_excel('assignment3_data/fruit/'+f)
    #set a blank dataframe to store information from the specific fruit in correct format
    f_info = pd.DataFrame()
    #extract information from original dataset
    f_info['price_per_lb']=[df.iloc[2,1]]
    f_info['yield']=[df.iloc[2,3]]
    f_info['lb_per_cup']=[df.iloc[2,4]]
    f_info['price_per_cup']=[df.iloc[2,6]]
    #finish dataframe for specific fruit with correct formating
    f_info['type']=['fruit']
    f_info['food']=[f[:-5]]
    f_info['form']=['Fresh1']
    #rearrange it with same column order as the fruit matrix
    f_info = f_info.reindex_axis(maincolumns, axis=1)
    #combine information of the specific fruit dataframe with fruit dataframe
    frames=[fruit,f_info]
    fruit=pd.concat(frames,ignore_index=True)

#there is one profile that does not contain price information fresh row, that need to drop na
fruit=fruit.dropna()
#the last file named "dated" is a file does not contain specific fruit information
#thus need to drop it as well
fruit=fruit.drop([28])
fruit
```

```python
import pandas as pd
import numpy as np
import os

#get files names from vegetable fold and store the file names 
files = os.listdir('assignment3_data/vegetables')

#set the column name for vegetable data frame
maincolumns = ['type','food','form','price_per_lb','yield','lb_per_cup','price_per_cup']

#set a blank dataframe for vegetable information
vegetable = pd.DataFrame(columns=maincolumns)


for f in files:
    df = pd.read_excel('assignment3_data/vegetables/'+f)
    f_info = pd.DataFrame()
    f_info['price_per_lb']=[df.iloc[2,1]]
    f_info['yield']=[df.iloc[2,3]]
    f_info['lb_per_cup']=[df.iloc[2,4]]
    f_info['price_per_cup']=[df.iloc[2,6]]
    f_info['type']=['vegetable']
    f_info['food']=[f[:-5]]
    f_info['form']=['Fresh1']
    f_info = f_info.reindex_axis(maincolumns, axis=1)
    frames=[vegetable,f_info]
    vegetable=pd.concat(frames,ignore_index=True)

vegetable
```

*  The previous code does not return the correct price for tomatoes

This is because in tomatoes file, the information about tomato is not stored in Fresh row, but the other rows. While as in the function, the information of food was extracted from Fresh row specificly, thus it's not working for tomatoes file. The other files have the same problems due the the same reason. 


```python
vegetable=vegetable.dropna()
vegetable[['price_per_lb','yield','lb_per_cup','price_per_cup']] = vegetable[['price_per_lb','yield','lb_per_cup','price_per_cup']].apply(pd.to_numeric)
fruit.dtypes==vegetable.dtypes
frames=[fruit,vegetable]
full=pd.concat(frames,ignore_index=True)
full
```

```python
expensive_f=fruit['food'][fruit['price_per_lb'].argmax()]
cheap_f=fruit['food'][fruit['price_per_lb'].argmin()]
print('The most expensive fruit is:\n',expensive_f,"\nThe cheapest fruit is:\n",cheap_f)
```

*  2 graph of price per lb for fruit and vegetable was drawn respectively
*  Notice that most fruit has a higher price range than vegetable: the highest price for fruit is around 7 and for food is around 4, the lowest price are both around 0, but the trends of decreasing have similar slope. 

```python
import matplotlib
import matplotlib.pyplot as plt
#store fruit information in a new dataframe and rearrange it 
#with the decreasing order of price_per_lb
price_f=fruit[['food','price_per_lb']]
price_f=price_f.sort_values(['price_per_lb'],ascending=False)
#reset the index
price_f= price_f.reset_index()
#plot graph 
price_f.plot(y='price_per_lb',title="fruit")
plt.show()

#store vegetable information in a new dataframe and rearrange it 
#with the decreasing order of price_per_lb
price_v=vegetable[['food','price_per_lb']]
price_v=price_v.sort_values(['price_per_lb'],ascending=False)
#reset the index
price_v= price_v.reset_index()
#plot graph
price_v.plot(y='price_per_lb',title="vegetable")
plt.show()
```

2.)Secondly, there is some inconsistancy in the dataset.

     However, the difference between calculated price per cup and real price per cup that has been recorded in the dataset, which was denoted as the real price per cup,  was noticable. As shown in the graph "difference between real and calculated price per cup",  the real price per cup is likely higher than the calculated price per cup, since the difference is higher than 0 in most of the time. 
     

```python
full_2=full
#get observations with highest yield value
full_2.sort_values(['yield'],ascending=False).head()
```

```python
full_1=full
#calculte price per cup by multiplying 2 columns
full_1['ppc']=full_1['price_per_lb']*full_1['lb_per_cup']
#get the difference between calculated data and the real data
full_1['diff']=full_1['price_per_cup']-full_1['ppc']
#extract information to plot graph 
full_1=full_1[['food','diff']]
full_1.plot(y='diff',title="difference between real and estimated price per cup")
plt.show()
```
